```python
r = 5
input_len = 4
#f_num = 2
x0, x1, x2, x3 = SR.var('x0 x1 x2 x3')
# for i in range(r):
#     for j in range(f_num):
#         SR.var('f_{0}_{1}'.format(j, i))

# SR.var('f_0_1 f_1_1')
# #F0 = function("F0")
# #F1 = function("F1")

# expr = f_0_1 + f_1_1 + x2
# print(expr.coefficients(expr.arguments()[0])[0][0])
# print(expr.coefficients(expr.arguments()[0]))
# print(expr.coefficients())
# print(expr.arguments()[0])
# print(expr.arguments())
# print("list = ", expr.list())
# if 'F' in str(expr.coefficients(expr.arguments()[0])[0][0]):
#     print(expr)
# #print(expr.coefficient(x1))
# print(expr.operands())
```

```python
def simp1(expr):
    #print("expr111 ===== ",expr)
    op=expr.operands()
    #print("op111 ===== ",op)
    for iter1 in op:
        #print("iter1 ===== ",iter1)
        arg = iter1.arguments()[0]
        #获取系数xs
        xs = int(iter1.coefficients(arg)[0][0])
        #print(iter1,", xishu= ",xs)
        if xs % 2 == 0 and xs != 0:
            expr = expr - iter1
        if xs % 2 == 1 and xs != 1:
            expr = expr - (xs-1)*arg

    #print(expr)
    return expr

def simp2(expr):
    #print("expr111 ===== ",expr)
    op=expr.operands()
    #print("op111 ===== ",op)
    for iter1 in op:
        #print("iter1 ===== ",iter1)
        arg = iter1.arguments()[0]
        #获取系数xs
        xs = int(iter1.coefficients(arg)[0][0])
        #print(iter1,", xishu= ",xs)
        if xs % 2 == 0 and xs != 0:
            expr = expr - (xs-1)*arg
        if xs % 2 == 1 and xs != 1:
            expr = expr - (xs-1)*arg

    #print(expr)
    return expr

#print("output[10] = ", output[10])
#print("=======!",simp1(2*x1 + x15 + 17*x2 + 3*x3 + x4 + 12*x5 + x6 + x7 + F1(x13 + x7 + F5(x2)) + F7(x0)))
#print("====!",simp1(2*x1 + x15 + 17*x2 + 3*x3 + x4 + 12*x5 + x6 + x7 + F1(x13 + x7 + F5(x2)) + F7(x0)))
```

```python
#f_i_j代表第j轮的轮函数f_i
for i in range(r):
    f1 = x0
    f2 = x1
    f3 = x2 + SR.var('f_0_{}'.format(i)) + x0
    f4 = x3 + SR.var('f_1_{}'.format(i)) + x1
    # linear layer
    #x2 = x2 + x0
    #x3 = x3 + x1
    # perm layer
    y0 = f3
    y1 = f4
    y2 = f1
    y3 = f2
    x0, x1, x2, x3 = y0, y1, y2, y3
    output = [y0, y1, y2, y3]

#print(" x1 = ", x1)
#print("!!=!",simp1(2*x1))
for i in range(input_len):
    output[i]=simp1(output[i])
    print("output[",i,"]= ", output[i])

#print(output[1])
#print(output[1].arguments())
#print(output[1].coefficients(output[1].arguments()[0]))

def clean_xi(expr):
    op=expr.operands()
    #print("op =", op)
    for iter1 in op:
        str_iter1 = str(iter1)
        if 'f' in str_iter1:
            continue
        else:
            expr = expr - iter1
    return expr

# for i in range(len(output)):
#     output[i] = clean_xi(output[i])
#print("output[",i,"]= ", output[i])
#print(clean_xi(output[1]))#对单个元素而言，存在debug
#print(clean_xi(output[0] + output[1]))
```

```python
#这里必须手动修改一下，看output的输出，手动写两个值到all_F中，不然执行all_F.operands()会报错
#simp2专门用来对全部的轮函数Fi求和
all_F = simp2(clean_xi(output[0] + output[1] + output[2] + output[3]))
#print(simp2(2*f_0_0 + f_0_1 + 2*f_1_0 + f_1_1))
#print("all_F1 = ", all_F)
def add_all(expr):
    expr_op=expr.operands()
    #print("expr_op = ", expr_op)
    for iter1 in expr_op:
        str_iter1 = str(iter1)
        #print("str_iter1 = ", str_iter1)
        if 'f' in str_iter1:
            global all_F
            all_F_op=all_F.operands()
            #print("all_F_op =", all_F_op)
            if iter1 not in all_F_op:
                #print(iter1, " not in ",all_F_op)
                all_F = all_F + iter1
        else:
            continue

#求所有Fi相加
#add all expression, calculate the max number of expr
for iter1 in output:
    #print("iter = ", iter1)
    add_all(iter1)

all_F=simp1(all_F)
print("all_F2 = ", all_F)

#clean_xi(output[0])
```

```python
def coeff1(expr, all):
    coeff1 = [0] * len(all)
    expr_op=expr.operands()
    all_op=all.operands()
    #print("all_op ===== ",all_op)
    #print("expr_op ===== ",expr_op)
    
    for iter_expr in expr_op:
        str_iter_expr = str(iter_expr)
        for iter_all in all_op:
            if str(iter_all) == str_iter_expr:
                #print("iter_all == ",iter_all)
                #print (str(iter_all), " == ", str_iter_expr, "\n")
                coeff1[all_op.index(iter_all)] = 1
            else:
                continue
    #print(expr)
    return coeff1

#m行n列的二维数组 L = [[0] * n] * m
coeff_matrix = [[0] * len(all_F)] * input_len
print("all_F = ", all_F)
print("len(all_F) = ", len(all_F))

i = 0
for iter1 in output:
    #print("output_iter = ", iter1)
    coeff_matrix[i] = coeff1(iter1, all_F)
    i = i + 1
#print("matrix = ", coeff_matrix)
```
