```python
round_number = 7
input_len = 7
x0, x1, x2, x3, x4, x5, x6 = SR.var('x0 x1 x2 x3 x4 x5 x6')

def simp1(expr):
    op=expr.operands()
    for iter1 in op:
        arg = iter1.arguments()[0]
        #get the coefficients
        xs = int(iter1.coefficients(arg)[0][0])
        #print(iter1,", xishu= ",xs)
        if xs % 2 == 0 and xs != 0:
            expr = expr - iter1
        if xs % 2 == 1 and xs != 1:
            expr = expr - (xs-1)*arg
    return expr

#simp2 adds all the round functions Fi
def simp2(expr):
    op=expr.operands()
    for iter1 in op:
        arg = iter1.arguments()[0]
        #get the coefficients
        xs = int(iter1.coefficients(arg)[0][0])
        if xs % 2 == 0 and xs != 0:
            expr = expr - (xs-1)*arg
        if xs % 2 == 1 and xs != 1:
            expr = expr - (xs-1)*arg
    return expr
```

```python
#f_i_j preresents the j-th round function f_i
for i in range(round_number):
    s0 = x0
    s1 = x1 + SR.var('f_0_{}'.format(i))
    s2 = x2
    s3 = x3
    s4 = x4
    s5 = x5
    s6 = x6
    # perm layer (2,3,4,5,1,6,0)
    y0 = s6
    y1 = s4
    y2 = s0
    y3 = s1
    y4 = s2
    y5 = s3
    y6 = s5
    x0, x1, x2, x3, x4, x5, x6 = y0, y1, y2, y3, y4, y5, y6
    output = [y0, y1, y2, y3, y4, y5, y6]

for i in range(input_len):
    print("output[",i,"]= ", output[i])
    output[i]=simp1(output[i])

def clean_xi(expr):
    op=expr.operands()
    for iter1 in op:
        str_iter1 = str(iter1)
        if 'f' in str_iter1:
            continue
        else:
            expr = expr - iter1
    return expr
```

```python
all_F = simp2(clean_xi(output[0] + output[1] + output[2] + output[3]))
def add_all(expr):
    expr_op=expr.operands()
    for iter1 in expr_op:
        str_iter1 = str(iter1)
        if 'f' in str_iter1:
            global all_F
            all_F_op=all_F.operands()
            if iter1 not in all_F_op:
                all_F = all_F + iter1
        else:
            continue

#add all expression, calculate the max number of expr
for iter1 in output:
    add_all(iter1)

all_F=simp1(all_F)
print("all_F = ", all_F)
```

```python
def coeff1(expr, all):
    coeff1 = [0] * len(all)
    expr_op=expr.operands()
    all_op=all.operands()
    for iter_expr in expr_op:
        str_iter_expr = str(iter_expr)
        for iter_all in all_op:
            if str(iter_all) == str_iter_expr:
                coeff1[all_op.index(iter_all)] = 1
            else:
                continue
    return coeff1

coeff_matrix = [[0] * len(all_F)] * input_len
i = 0
for iter1 in output:
    coeff_matrix[i] = coeff1(iter1, all_F)
    i = i + 1

A = Matrix(coeff_matrix)
print(A)
A.rank()
```
