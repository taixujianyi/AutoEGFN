```python
round_number = 17
input_len = 32
x0, x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11, x12, x13, x14, x15 = SR.var('x0 x1 x2 x3 x4 x5 x6 x7 x8 x9 x10 x11 x12 x13 x14 x15')
x16, x17, x18, x19, x20, x21, x22, x23, x24, x25, x26, x27, x28, x29, x30, x31 = SR.var('x16 x17 x18 x19 x20 x21 x22 x23 x24 x25 x26 x27 x28 x29 x30 x31')
def simp1(expr):
    op=expr.operands()
    for iter1 in op:
        arg = iter1.arguments()[0]
        #get the coefficients
        xs = int(iter1.coefficients(arg)[0][0])
        if xs % 2 == 0 and xs != 0:
            expr = expr - iter1
        if xs % 2 == 1 and xs != 1:
            expr = expr - (xs-1)*arg
    return expr

#simp2 adds all the round functions Fi
def simp2(expr):
    op=expr.operands()
    for iter1 in op:
        arg = iter1.arguments()[0]
        #get the coefficients
        xs = int(iter1.coefficients(arg)[0][0])
        if xs % 2 == 0 and xs != 0:
            expr = expr - (xs-1)*arg
        if xs % 2 == 1 and xs != 1:
            expr = expr - (xs-1)*arg
    return expr
```

```python
#f_i_j preresents the j-th round function f_i
for i in range(round_number):
    s0 = x0 + SR.var('f_0_{}'.format(i))
    s1 = x1 + SR.var('f_1_{}'.format(i))
    s2 = x2 + SR.var('f_2_{}'.format(i))
    s3 = x3 + SR.var('f_3_{}'.format(i))
    s4 = x4 + SR.var('f_4_{}'.format(i))
    s5 = x5 + SR.var('f_5_{}'.format(i))
    s6 = x6 + SR.var('f_6_{}'.format(i))
    s7 = x7 + SR.var('f_7_{}'.format(i))
    s8 = x8 + SR.var('f_8_{}'.format(i))
    s9 = x9 + SR.var('f_9_{}'.format(i))
    s10 = x10 + SR.var('f_10_{}'.format(i))
    s11 = x11 + SR.var('f_11_{}'.format(i))
    s12 = x12 + SR.var('f_12_{}'.format(i))
    s13 = x13 + SR.var('f_13_{}'.format(i))
    s14 = x14 + SR.var('f_14_{}'.format(i))
    s15 = x15 + SR.var('f_15_{}'.format(i))
    s16 = x16
    s17 = x17
    s18 = x18
    s19 = x19
    s20 = x20
    s21 = x21
    s22 = x22
    s23 = x23
    s24 = x24
    s25 = x25
    s26 = x26
    s27 = x27
    s28 = x28
    s29 = x29
    s30 = x30
    s31 = x31
    # perm layer
    y0 = s31
    y1 = s0
    y2 = s1
    y3 = s2
    y4 = s3
    y5 = s4
    y6 = s5
    y7 = s6
    y8 = s7
    y9 = s8
    y10 = s9
    y11 = s10
    y12 = s11
    y13 = s12
    y14 = s13
    y15 = s14
    y16 = s15
    y17 = s16
    y18 = s17
    y19 = s18
    y20 = s19
    y21 = s20
    y22 = s21
    y23 = s22
    y24 = s23
    y25 = s24
    y26 = s25
    y27 = s26
    y28 = s27
    y29 = s28
    y30 = s29
    y31 = s30
    x0, x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11, x12, x13, x14, x15 = y0, y1, y2, y3, y4, y5, y6, y7, y8, y9, y10, y11, y12, y13, y14, y15
    x16, x17, x18, x19, x20, x21, x22, x23, x24, x25, x26, x27, x28, x29, x30, x31 = y16, y17, y18, y19, y20, y21, y22, y23, y24, y25, y26, y27, y28, y29, y30, y31
    output = [y0, y1, y2, y3, y4, y5, y6, y7, y8, y9, y10, y11, y12, y13, y14, y15, y16, y17, y18, y19, y20, y21, y22, y23, y24, y25, y26, y27, y28, y29, y30, y31]

for i in range(input_len):
    print("output[",i,"]= ", output[i])
    output[i]=simp1(output[i])

def clean_xi(expr):
    op=expr.operands()
    for iter1 in op:
        str_iter1 = str(iter1)
        if 'f' in str_iter1:
            continue
        else:
            expr = expr - iter1
    return expr
```

```python
all_F = simp2(clean_xi(output[0] + output[1] + output[2] + output[3]))
def add_all(expr):
    expr_op=expr.operands()
    for iter1 in expr_op:
        str_iter1 = str(iter1)
        if 'f' in str_iter1:
            global all_F
            all_F_op=all_F.operands()
            if iter1 not in all_F_op:
                all_F = all_F + iter1
        else:
            continue

#add all expression, calculate the max number of expr
for iter1 in output:
    add_all(iter1)

all_F=simp1(all_F)
print("all_F = ", all_F)
```
