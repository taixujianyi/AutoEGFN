```python
round_number = 3
input_len = 22
x0, x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11, x12, x13, x14, x15 = SR.var('x0 x1 x2 x3 x4 x5 x6 x7 x8 x9 x10 x11 x12 x13 x14 x15')
x16, x17, x18, x19, x20, x21 = SR.var('x16 x17 x18 x19 x20 x21')
def simp1(expr):
    op=expr.operands()
    for iter1 in op:
        arg = iter1.arguments()[0]
        #get the coefficients
        xs = int(iter1.coefficients(arg)[0][0])
        if xs % 2 == 0 and xs != 0:
            expr = expr - iter1
        if xs % 2 == 1 and xs != 1:
            expr = expr - (xs-1)*arg
    return expr

#simp2 adds all the round functions Fi
def simp2(expr):
    op=expr.operands()
    for iter1 in op:
        arg = iter1.arguments()[0]
        #get the coefficients
        xs = int(iter1.coefficients(arg)[0][0])
        if xs % 2 == 0 and xs != 0:
            expr = expr - (xs-1)*arg
        if xs % 2 == 1 and xs != 1:
            expr = expr - (xs-1)*arg
    return expr
```

```python
all_F = simp2(clean_xi(output[0] + output[1] + output[2] + output[3]))
def add_all(expr):
    expr_op=expr.operands()
    for iter1 in expr_op:
        str_iter1 = str(iter1)
        if 'f' in str_iter1:
            global all_F
            all_F_op=all_F.operands()
            if iter1 not in all_F_op:
                all_F = all_F + iter1
        else:
            continue

#add all expression, calculate the max number of expr
for iter1 in output:
    add_all(iter1)

all_F=simp1(all_F)
print("all_F = ", all_F)
```

```python
def coeff1(expr, all):
    coeff1 = [0] * len(all)
    expr_op=expr.operands()
    all_op=all.operands()
    for iter_expr in expr_op:
        str_iter_expr = str(iter_expr)
        for iter_all in all_op:
            if str(iter_all) == str_iter_expr:
                coeff1[all_op.index(iter_all)] = 1
            else:
                continue
    return coeff1

coeff_matrix = [[0] * len(all_F)] * input_len
i = 0
for iter1 in output:
    coeff_matrix[i] = coeff1(iter1, all_F)
    i = i + 1

A = Matrix(coeff_matrix)
print(A)
A.rank()
```
